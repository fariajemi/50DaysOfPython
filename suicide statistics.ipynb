```python
subdf = df[['country','year','sex','suicides_no']]
subdf.head()
```

```python
sex_dummies = pd.get_dummies(subdf.sex, prefix="sex")
```

```python

df_with_dummies = pd.concat([subdf,sex_dummies],axis='columns')
```

```python
df_with_dummies.head()
```

```python
df_with_dummies.drop('sex',axis='columns',inplace=True)
df_with_dummies.head()
```

```python
X = df_with_dummies
X.head()
```

```python
df_with_dummies.drop('country',axis='columns',inplace=True)
df_with_dummies.head()
```

```python
X = df_with_dummies
X.head()
```

```python

y = df.suicides_no
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```
